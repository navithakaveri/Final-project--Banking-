# DATA PREPROCESSING

```python
#HANDLING MISSING VALUES
df.isnull().sum()
```

```python
#REMOVING DUPLICATES 
duplicates=df.duplicated()
duplicates
```

```python
#categorical and numerical data
import numpy as np
categorical_data=df.select_dtypes(include=[object]).columns
numerical_cols=df.select_dtypes(include=np.number).columns.tolist()

print("categorical datas:",categorical_data)

print("numerical datas:",numerical_cols)
```

```python
#HANDLING OUTLIERS-IDENTIFYING OUTLIERS
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR))).sum()
print("Number of outliers:\n", outliers)
```

```python
#ploted the outlier using histogram and box plot
for col in numerical_cols:
    print(f'Skew for {col}:', round(df[col].skew(), 2))
    plt.figure(figsize=(14, 6))
    
    # Histogram
    plt.subplot(1, 2, 1)
    df[col].hist(grid=False, color='pink')
    plt.title(f'Histogram of {col} (Original)')
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot of {col} (Original)')
    
    plt.show()
```

```python
# Calculated IQR and bounds for each numerical column
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Defined lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers for each numerical column separately
df_capped = df.copy()
for col in numerical_cols:
    df_capped[col] = df_capped[col].clip(lower=lower_bound[col], upper=upper_bound[col])

# Display the result
df_capped.head(2)
```

```python
Q1 =   df_capped[numerical_cols].quantile(0.25)
Q3 =   df_capped [numerical_cols].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df_capped [numerical_cols] < (Q1 - 1.5 * IQR)) | (df_capped [numerical_cols] > (Q3 + 1.5 * IQR))).sum()
print("Number of outliers:\n", outliers)
```

```python
for col in numerical_cols:
    print(f'Skew for {col}:', round(df[col].skew(), 2))
    plt.figure(figsize=(14, 6))
    
    # Original data
    plt.subplot(1, 2, 1)
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot of {col} (Original)')
    
    # Capped data
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df_capped[col], color='lightgreen')
    plt.title(f'Boxplot of {col} (Capped)')
    
    plt.show()
```

```python
df=pd.DataFrame(data)
```

```python
#droping columns which is unique and it doesn't needed for model training
df1 = df.drop([ 'ID','Customer_ID', 'Name','Month','SSN','Type_of_Loan'],axis=1)
df1.info()
```

```python
#identifying categorical columns for encoding
cat_cols1 = list(df1.select_dtypes(include='object'))
cat_cols1
```

```python
df1['Occupation'].unique()
```

```python
df1['Credit_Mix'].unique()
```

```python
df1['Payment_of_Min_Amount'].unique()
```

```python
df1['Payment_Behaviour'].unique()
```

```python
# encoding categorical data using ordinal encoder
Credit_Mix_map = { 'Bad':0, 'Standard':1, 'Good': 2  }
Payment_of_Min_Amount_map = { 'No':0, 'NM':1, 'Yes':2 }
Payment_Behaviour_map = { 'Low_spent_Small_value_payments': 0 , 'Low_spent_Medium_value_payments':1, 'Low_spent_Large_value_payments':2,
                         'High_spent_Small_value_payments':3, 'High_spent_Medium_value_payments':4, 'High_spent_Large_value_payments':5
                        }

Occupation_map ={'Scientist':0, 'Teacher':1, 'Engineer':2, 'Entrepreneur':3, 'Lawyer':4,
       'Media_Manager':5, 'Doctor':6, 'Journalist':7, 'Manager':8, 'Musician':9,
       'Mechanic':10, 'Writer':11, 'Accountant':12, 'Architect':13, 'Developer':14}
```

```python
df1['Credit_Mix'] = df1['Credit_Mix'].map(Credit_Mix_map)
df1['Occupation']=df1['Occupation'].map(Occupation_map)

df1['Payment_of_Min_Amount'] = df1['Payment_of_Min_Amount'].map(Payment_of_Min_Amount_map)

df1['Payment_Behaviour'] = df1['Payment_Behaviour'].map(Payment_Behaviour_map)
```
